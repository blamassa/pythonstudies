# ERROR BAR CHARTS

```python
x = np.arange(0,4,0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x, y, yerr=e1, fmt='.-');
```

obs: formatting arguments
- `fmt`: plot format of lines. (default = None)
- `ecolor`: Cor das barras de erros
- `elinewidth`: Largura das barras de erros (default = mesma que a line width)
- `capsize`: Especifica o tamanho da capsize das barras de erro (barrinhas de limite)

```python
e2 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x, y, yerr=e1, xerr=e2, fmt='.-', capsize=2.5);
```

# BAR CHARTS
Grfico de barras (horizontal ou vertical). Geralmente usados para comparar 2 ou mais valores.

### Vertical Bars

```python
plt.bar([1,2,3],[3,2,5])
```

obs: argumentos
- `width`: largura das barras (default = 0.8)
- `color`
- `xerr, yerr`: barras de erros no bar chart (raramente utilizado)
- `bottom`: bottom cordinates of the bars

```python
plt.bar([1,2,3],[3,2,5], width = 0.1)
```

**Exemplo:**

```python
data1 = 10*np.random.rand(5)
data2 = 10*np.random.rand(5)
data3 = 10*np.random.rand(5)

e2 = 0.5 * np.abs(np.random.randn(len(data2)))
locs = np.arange(1, len(data1)+1)
width = 0.27

plt.bar(locs, data1, width=width)
plt.bar(locs+width, data2, yerr=e2, width=width,
color='red')
plt.bar(locs+2*width, data3, width=width, color='green')
plt.xticks(locs + width*1.5, locs)
```

# PIE CHARTS

```python
# importante definir o tamanho da figura como quadrado, se no vira uma elipse
plt.figure(figsize=(5,5))
# se a soma no for 100 ou 1, vai ser uma pizza irregular
x = [45, 35, 20]

labels = ['Cats', 'Dogs', 'Fishes']
plt.pie(x, labels=labels)
```

obs: argumentos (aka kwargs**)

- `explode`: array of the same lenght of x. Cada valor especifica a frao do raio que o pedao se afasta do centro da pizza
- `colors`
- `labels, labeldistance`: lista do mesmo tamanho que x
- `shadow`
- `autopct`: printa a porcentagem. Necessita a seguinte formatting string: autopct='%1.1f%%'

```python
plt.figure(figsize=(5,5))
x = [4, 9, 21, 55, 30, 18]
labels = ['Swiss', 'Austria', 'Spain', 'Italy', 'France',
'Benelux']
explode = [0.2, 0.1, 0, 0, 0.1, 0]
plt.pie(x, labels=labels, explode=explode, autopct='%1.1f%%');
```

# SCATTER PLOTS
Plotagem em pontos, e no linhas

```python
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.scatter(x,y)
```

obs: kwargs**

- `s`: size of markers in pixel * pixel. Pode ser um valor nico (todos os markers so do mesmo tamanho) ou uma array do mesmo tamanho de X e Y (cada ponto tem seu prprio tamanho)
- `c`: color. Pode ser valor nico ou uma lista de cores do mesmo tamanho que X e Y. Pode ser nmeros ou strings de cores,
- `marker`: formato do marker. ex s = square, o = circle, x = cross, etc

```python
size = 50*np.random.randn(1000) # definindo uma lista de tamanhos
colors = 50*np.random.randn(1000)
plt.scatter(x, y, s=size, c=colors)
```
