## Multiline Plots

```python
x = range(1,5)
plt.plot(x, [xi*1.5 for xi in x])
plt.plot(x, [xi*3.0 for xi in x])
plt.plot(x, [xi/3.0 for xi in x])
```

```python
x = range(1,5)
plt.plot(x, [xi*1.5 for xi in x], x, [xi*3.0 for xi in x], x,
[xi/3.0 for xi in x])
```

### obs: usar o metodo arange do NumPy facilita muito mais que usar simplismente a lista gerada pelo range()

```python
x = np.arange(1,5)
plt.plot(x,x*1.5,x,x*3,x,x/3)
```

obs: np.arange(x) simula o range(), mas retorna uma array

## Grid, axes and labels

### Grid

```python
x = np.arange(1,5)
plt.plot(x,x*1.5,x,x*3,x,x/3)
plt.grid(True)
```

### Axes (definindo os limites dos eixos)

```python
x = np.arange(1,5)
plt.plot(x,x*1.5,x,x*3,x,x/3)
# o comando plt.axis() printa os limites atuais
plt.axis([0,5,-1,13]) # set os novos limites
```

### Labels

```python
plt.plot([1,3,2,4])
plt.xlabel('This is X axis')
plt.ylabel('This is Y axis')
```

## Titles and Legends

```python
plt.plot([1,3,2,4])
plt.title('Simple plot')
```

```python
x = np.arange(1,5)
plt.plot(x,x*1.5, label='Normal')
plt.plot(x,x*3.0, label='Fast')
plt.plot(x,x/3.0, label='Slow')
plt.legend()
```
